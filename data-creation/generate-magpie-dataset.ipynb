```python
from distilabel.llms import TransformersLLM
from distilabel.steps.tasks import MagpieGenerator
from distilabel.pipeline import Pipeline
from distilabel.steps.tasks import MagpieGenerator, TextGeneration

USER_INSTRUCTIONS_PROMPT = (
    "You are an AI assistant designed to help users with their questions. You specialize in providing accurate information for all things outdoors. ",
    "People bring you on hikes, camping trips, and other outdoor adventures to help them identify plants, animals, and other things they encounter. ",
    "You also provide in depth survival tips on things like how to find water, first aid, and how to build a shelter. ",
    "Questions will be about all kinds of environments, from forests to deserts to mountains. ",
)

ASSISTANT_INSTRUCTIONS_PROMPT = (
    "You are an AI assistant designed to help users with their questions. You specialize in providing accurate information for all things outdoors. ",
    "People bring you on hikes, camping trips, and other outdoor adventures to help them identify plants, animals, and other things they encounter. ",
    "You also provide in depth survival tips on things like how to find water, first aid, and how to build a shelter. ",
    "Questions will be about all kinds of environments, from forests to deserts to mountains. ",
)
```

```python
with Pipeline(name="hiker-info-tuning") as pipeline:
#    device = 'cuda'
    generate_instructions = MagpieGenerator(
#        llm=OpenAILLM(api_key=API_KEY,
#                base_url="https://api.together.xyz/v1",
#                model="meta-llama/Meta-Llama-3.1-70B-Instruct-Turbo",
#        )     ,
        llm=TransformersLLM(
            model="meta-llama/Meta-Llama-3-8B-Instruct",
            magpie_pre_query_template="llama3",
            generation_kwargs={
                "temperature": 1.0,
                "max_new_tokens": 256,
            },
            device="cuda",
        ),
        system_prompt=USER_INSTRUCTIONS_PROMPT,
        batch_size=1,
        num_rows=1,
        only_instruction=True,
    )

    response = TextGeneration(
     #   llm=OpenAILLM(api_key=API_KEY,
     #           base_url="https://api.together.xyz/v1",
     #           model="meta-llama/Meta-Llama-3.1-70B-Instruct-Turbo",
        #    generation_kwargs={"temperature": 0.8, "max_new_tokens": 2048},
    #    )     ,
 #       llm=InferenceEndpointsLLM(
#            model_id="meta-llama/Meta-Llama-3.1-70B-Instruct",
#            tokenizer_id="meta-llama/Meta-Llama-3.1-70B-Instruct",
#            magpie_pre_query_template="llama3",
#            generation_kwargs={"temperature": 0.8, "max_new_tokens": 2048},
#        ),    
        llm=TransformersLLM(
            model="meta-llama/Meta-Llama-3-8B-Instruct",
         #   magpie_pre_query_template="llama3",
            generation_kwargs={
                "temperature": 1.0,
                "max_new_tokens": 256,
            },
        #    device="cuda",
        )    ,
#        system_prompt=ASSISTANT_INSTRUCTIONS_PROMPT,
        input_batch_size=1,
    )
 #   device = "cuda"# if torch.cuda.is_available() else "cpu"
    generate_instructions >> response

#distiset = pipeline.run()
```
